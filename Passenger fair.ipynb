```python
class Passenger:
    
    def __init__(self,passengerName, passengerAge, passengerTraveled):
        
        self.Name = passengerName
        self.Age = passengerAge
        self.Travel = passengerTraveled
        
    def calculateTicketFare(list_Passenger,fare):
        total = 0
        for i in list_Passenger:
            
            if (i.Age >=60 or i.Age<12):
                total += (i.Travel * fare)
                
            elif (i.Age >20 and i.Age<60 ):
                total += ((i.Travel * fare) + ((i.Travel * fare) * (12/100)))
            
            elif (i.Age >=12 and i.Age <=20):
                total += ((i.Travel * fare) + ((i.Travel * fare) * (10/100)))
                
        return total
    
    
    def countSeniorJuniorPassengers():
        count = 0
        for i in list_Passenger:
            
            if (i.Age >=60) or (i.Age<12):
                count += 1
                
        return count
            
if __name__ == "__main__":
    
    N = int(input())
    list_Passenger = []
    for i in range(N):
        name = input()
        age = int(input())
        travel = int(input())
        list_Passenger.append(Passenger(name,age,travel))
    
    fare = int(input())
    
    print(Passenger.calculateTicketFare(list_Passenger,fare))
    print(Passenger.countSeniorJuniorPassengers())
```
