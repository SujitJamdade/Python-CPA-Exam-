```python
class Movie:
    
    def __init__(self,id,movie_name,ticket_cost):
        
        self.id = id
        self.movie_name = movie_name
        self.ticket_cost = ticket_cost
    
    def priceCategory(self,category= "Default"):
           
        if self.ticket_cost > 0 and self.ticket_cost < 150:
            return "General"
        elif self.ticket_cost >= 150 and self.ticket_cost < 250:
            return "Silver"
        elif self.ticket_cost >= 250 and self.ticket_cost < 350:
            return "Gold"
        else :
            return "Platinum"
        
class Movie_Ticket:
    
    def __init__(self,customer_name,customer_movie,view_count):
        
        self.customer_name = customer_name
        self.customer_movie = customer_movie
        self.view_count = view_count
        
    
def getcategoryWiseCount(movies_list):
    
    category_dict = {}  
    for i in movies_list:
        
        category_ticket = i.priceCategory()
        
        if category_ticket in category_dict:
            
            category_dict[category_ticket] += 1
            
        else:
            category_dict[category_ticket] = 1
            
    return category_dict

def bookMovieTicket(movies_list, customer_name, customer_movie,view_count):
    
    bookMovie = [] #[Ramesh, 500]
    for i in movies_list:
        
        if customer_movie == i.movie_name:
            price = i.ticket_cost *view_count
            bookMovie.append(customer_name)
            bookMovie.append(price)
            
            return bookMovie
        
    return None
        
if __name__ == "__main__":
    
    movies_no = int(input())
    movies_list = []
    
    for i in range(movies_no):
        id = int(input())
        movie_name = input()
        ticket_cost = int(input())
        movies_list.append(Movie(id,movie_name,ticket_cost))
        
        
    customer_name = input()
    customer_movie = input()
    view_count = int(input())
    
    obj = Movie_Ticket(customer_name,customer_movie,view_count)
    
    print("Category wise ticket count:")
    result1 = getcategoryWiseCount(movies_list)
    
    for cat, count in result1.items():
        print("{}:{}".format(cat, count))
        
    result2 = bookMovieTicket(movies_list, customer_name, customer_movie,view_count)
  
    if result2 == None:
        print("None")
    else:
        print(result2[0],result2[1])
```
