```python
class Blood:
    def __init__(self,bloodgroup,unitlnHand):
        self.bgroup=bloodgroup
        self.unit=unitlnHand

class BloodBank:
    def __init__(self,blis):
        self.blis=blis

    def isBloodAvailable(self,reqblood,requnit):
        for i in self.blis:
            if i.bgroup==reqblood and i.unit>=requnit:
                return "Blood Available"
        return "Blood Not Available"


    def findMinBloodStock(self):
        bloodunit=[]
        for i in self.blis:
            bloodunit.append(i.unit)
        minblood=min(bloodunit)
        blood=[]
        for i in self.blis:
            if i.unit==minblood:
                blood.append(i.bgroup)
        return blood

if __name__=="__main__":
    n=int(input())
    blis=[]
    for i in range(n):
        bgroup=input()
        unit=int(input())
        t1=Blood(bgroup,unit)
        blis.append(t1)

    reqblood=input()
    requnit=int(input())

    b=BloodBank(blis)
    res=b.isBloodAvailable(reqblood,requnit)
    print(res)
    res1=b.findMinBloodStock()
    for i in res1:
        print(i)
```
