```python
class Recipe:
    def __init__(self,id,name,ing,typ,calo):
        self.id = id
        self.name = name
        self.ing = ing
        self.typ = typ
        self.calo = calo
    
class Mykitchen:
    def __init__(self,rlist,klist):
        self.rlist = rlist
        self.klist = klist
        
    def getAvailableRecipeCount(self,rtype):
        c = 0
        for i in self.rlist:
            if rtype == i.typ:
                count = 0
                for j in klist:
                    if j in i.ing:
                        count += 1
                        
                        if count == 3:
                            c +=1
        
        if c == 0:
            print("major ingredients for given recipe type not found")
        else:
            print(c)
            
                        
                
    def getTwoRecipeWithLeastCalories(self):
        
        cal_list = []
        for i in self.rlist:
            cal_list.append(i.calo)
                
        cal_list.sort()
    
        result = []
        for i in range(2):
            for j in self.rlist:
                if cal_list[i] == j.calo:
                    result.append(j.id)
                    result.append(j.name)
                    result.append(j.typ)
                    result.append(j.calo)
                    
        return result
                    

    
if __name__ =="__main__":
    n = int(input())
    l = []
    for i in range(n):
        id = int(input())
        name = input()
        typ = input()
        ing_list =[]
        for i in range(3):
            ing=input()
            ing_list.append(ing)
        calo = float(input())        
        l.append(Recipe(id,name,ing_list,typ,calo))
        
    N = int(input())
    klist =[]
    for i in range(N):
        k = input()
        klist.append(k)
        
    rtype= input()
    
    obj = Mykitchen(l,klist)
    
    obj.getAvailableRecipeCount(rtype)
       
    res2 = obj.getTwoRecipeWithLeastCalories()
    
    for i in range(len(res2)):
        print(res2[i])
```
