```python
class Apparel:
    
    def __init__(self,apparelBrand, brandType, brandPrice, inStock):
        
        self.brand = apparelBrand
        self.type = brandType
        self.price = brandPrice
        self.inStock = inStock
        
        
class Store:
    
    def __init__(self,obj_list):
        self.list = obj_list
        
    def checkApparelAvailability(self,user_brand,user_type,user_size,user_req):
        check = 0
        for i in self.list:
          
            if user_brand == i.brand:
                if user_type == i.type:
                    if user_size in i.inStock:
                        if user_req <= i.inStock[user_size]:
                            check = 1
                            i.inStock[user_size] -= user_req
                            return i.inStock
        return check
            
if __name__ == "__main__":
    
    N = int(input())
    obj_list = []
    for i in range(N):
        br = input()
        ty = input()
        pri = int(input())
        n = int(input())
        inStock = {}
        for stock in range(n):
            size = input()
            availabel = int(input())
            inStock.update({size:availabel})
        
        obj_list.append(Apparel(br,ty,pri,inStock))
            
    user_brand = input()
    user_type = input()
    user_size = input()
    user_req = int(input())

    obj = Store(obj_list) 
            
    result = obj.checkApparelAvailability(user_brand,user_type, user_size, user_req)

    if result == 0:
        print("Size is not available")
        print("Apparel not found")
    else:
        print("Size is Available")
        for key,val in result.items():
            print("{}:{}".format(key,val))
```
