```python
class GoldInvoice:
    
    def __init__(self, item_id, item_name, item_qty, item_rate, item_weight, item_pwc,item_pdis):
        self.id = item_id
        self.name = item_name
        self.qty = item_qty
        self.rate = item_rate
        self.weight = item_weight
        self.pwc = item_pwc
        self.pdis = item_pdis
        
    def calc_item_price_exGST(lst,id_choice):
        
        for i in lst:            
            if i.id == id_choice:                
                item_cost = i.qty * i.rate * i.weight
                weight_percentage = item_cost * (i.pwc/100) 
                discount_percentage = item_cost * (i.pdis/100)                
                result = item_cost + weight_percentage - discount_percentage                
                return result
    
    def calc_item_price_inGST(lst,id_choice):
        
        for i in lst:            
            if i.id == id_choice:                
                item_cost = i.qty * i.rate * i.weight
                gst = item_cost * (3/100)
                weight_percentage = item_cost * (i.pwc/100) 
                discount_percentage = item_cost * (i.pdis/100)               
                result = item_cost + weight_percentage + gst - discount_percentage                
                return result  
        
        
if __name__ == "__main__":
    
    N = int(input())
    lst = []
    for i in range(N):
        iid = int(input())
        iname = input()
        iqty = float(input())
        irate = float(input())
        iweight = float(input())
        ipwc = float(input())
        ipdis = float(input())
        lst.append(GoldInvoice(iid,iname,iqty,irate,iweight,ipwc,ipdis))
        
    id_choice = int(input())    
    result1 = GoldInvoice.calc_item_price_exGST(lst,id_choice)    
    print("{:.3f}".format(result1))
    
    result2 = GoldInvoice.calc_item_price_inGST(lst,id_choice)    
    print("{:.2f}".format(result2))
```
