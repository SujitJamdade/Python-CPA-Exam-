```python
class Vehicle:
    def __init__(self, number, price, name):
        self.num=number
        self.price=price
        self.name=name
        
class Store:    
    def __init__(self, sN, vL):
        self.sN=sN
        self.vL=vL        
        
    def findMinumumVehicleByPrice(self):
        
        minPrice=min(self.vL, key=lambda x:x.price)
        return minPrice
    
    def findMaximumVehicleByNumber(self):
        
        maxnumbers =max(self.vL, key=lambda x:x.num)        
        return maxnumbers
        
if __name__ == "__main__":
    
    N= int(input())
    vL=[]
    for i in range(N):        
        n=int(input())
        p=int(input())
        name=str(input())                
        vL.append(Vehicle(n,p,name))
        
        
    obj= Store("XYZ", vL)
    
    ans=obj.findMinumumVehicleByPrice()
    
    if ans==None:
        print("No Data Found")
    else:
        print(ans.num)
        print(ans.price)
        print(ans.name)
        
        
    ans1=obj.findMaximumVehicleByNumber()
    
    if ans1==None:
        print("No Data Found")
    else:
        print(ans1.num)
        print(ans1.price)
        print(ans1.name)
```
