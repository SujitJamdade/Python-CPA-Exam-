```python
class Employee:
    def __init__(self, empName, design, salary, leaveB):
        self.n = empName
        self.d = design
        self.s = salary
        self.leaveB = leaveB


class Organization:
    def __init__(self, lis):
        self.lis = lis

    def checkLeaveEligibility(self, name, leaveType, days):
        demo = {}
        for i in self.lis:
            if i.n == name:
                if leaveType in i.leaveB:
                    if days <= i.leaveB[leaveType]:
                        i.leaveB[leaveType] -= days
                        for k in i.leaveB:
                            demo[k] = i.leaveB[k]
                        return demo
        return demo


if __name__ == "__main__":
    y = int(input())
    lis = []
    for i in range(y):
        n = input()
        d = input()
        s = int(input())
        N = int(input())
        dic = {}
        for i in range(N):
            leaveT = input()
            num = int(input())
            dic[leaveT] = num
        lis.append(Employee(n, d, s, dic))

    t1 = Organization(lis)
    name = input()
    leaveType = input().upper()
    days = int(input())
    res = t1.checkLeaveEligibility(name, leaveType, days)
    if not res:
        print("leave not granted")
        print("Employee not found")
    else:
        print("leave granted")
        for i, j in res.items():
            print("{} : {}".format(i, j))
```
