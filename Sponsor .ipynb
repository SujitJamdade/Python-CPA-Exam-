```python
class Sponsor:
    
    def __init__(self, sponserName, sponserCompany, subsidiaries, sponserCategory):
        self.name = sponserName
        self.company = sponserCompany
        self.sub = subsidiaries
        self.category = sponserCategory
        
def getSponsor(lst,spon_category):
    
    res = []
    count = 0
    for i in lst:
        if spon_category == i.category:
            res.append(i.name)
            count = 1
        
    if count == 0:
        print("No matching record found.")
    else:
        res.sort()
        for i in res:
            print(i)
        

def findSponsorWithMaximumSubsidiaries(lst):
    
    lst.sort(key=lambda x: len(x.sub), reverse = True)
    if len(lst)>0:
        return lst[0].name
    return None
        

        
if __name__ == "__main__":
    
    N = int(input())
    lst = []
    for i in range(N):
        n = input()
        c = input()
        sub_no = int(input())
        sub_list =[]
        for j in range(sub_no):
            sub_sid = input()
            sub_list.append(sub_sid)
            
        cate = input()
        
        lst.append(Sponsor(n,c,sub_list,cate))
        
    spon_category = input()
    getSponsor(lst, spon_category)  
    print(findSponsorWithMaximumSubsidiaries(lst))
```
