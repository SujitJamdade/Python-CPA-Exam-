```python
class Citizen:
    def __init__(self,Id,name,age,prof,war,city):
        self.Id = Id
        self.name = name
        self.age = age 
        self.prof = prof
        self.war = war
        self.city = city
        
class HealthCenter:
    
    def getHighestCity(self,l):
       
        
        d1 = {}  # for city : war
        d2 = {}  # for city : count 
        lst = [] 
        
        for i in l:
            if i.city in d1.keys():
                if i.war == "True":
                    d1[i.city] += 1
            
            else:
                if i.war == "True":
                    d1[i.city] = 1

       
        d1_keys = list(d1.keys())
        d1_values = list(d1.values())
         
        
        v = max(d1_values)
        ind = d1_values.index(v)
        
        k = d1_keys[ind]
        
        temp = {k:v}
        lst.append(temp)
        
        
        for i in l:
            if i.city in d2.keys():
                if i.age >=60:
                    d2[i.city] += 1
            
            else:
                if i.age >=60:
                    d2[i.city] = 1
                    
       
        d2_keys = list(d2.keys())
        d2_values = list(d2.values())
        
        v = max(d2_values)
        ind = d2_values.index(v)
        
        k = d2_keys[ind]
        
        temp = {k:v}
        lst.append(temp)
        
        return lst
        
        
    
    def cityWiseRequirement(self,l):
        
        D = {}
        
        for i in l:
            if i.city in D.keys():
            
                if i.war == "True":
                    
                    D[i.city][0] +=1
                    
                if i.age >= 60:
                    D[i.city][1] += 1
                
            else:
                D[i.city] = []
                if i.war == "True":
                    D[i.city].append(1)

                else:
                    D[i.city].append(0)

                if i.age >= 60:
                    D[i.city].append(1)
                else:
                    D[i.city].append(0)
            
        D_sorted = sorted(D.keys())
        for k in D_sorted:
            print(k, D[k])
                
    
if __name__ == "__main__":
    N = int(input())
    l = []
    for i in range(N):
        Id = int(input())
        name = input()
        age = int(input())
        prof = input()
        war = input()
        city = input()
        l.append(Citizen(Id,name,age,prof,war,city))
        
    obj = HealthCenter()
    
    res1 = obj.getHighestCity(l)
    
    for i in res1:
        print(i)
    
    obj.cityWiseRequirement(l)
    
```
