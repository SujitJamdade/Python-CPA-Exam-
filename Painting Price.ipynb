Creat a Painting class






```python
class Painting:
    def __init__(self, paintingID, paintingName, paintingPrice, paintingType):
        self.paintingId = paintingID
        self.paintingName = paintingName
        self.paintingPrice = paintingPrice
        self.paintingType = paintingType
        
class ShowRoom:
    def __init__(self, painting_List):
        self.pl = painting_List
        
    def getTotalPaintingPrice(self, pType):
        price = 0
        count = 0
        for i in self.pl:
            if i.paintingType == pType:
                price += i.paintingPrice
                count = 1
        if count == 0:
            return None
        else:
            return price
    
    def getMaxPainterName(self):
        painters= {}
        painter_list = []
        for i in self.pl:
            painter_list.append(i.paintingName)
            
        for i in painter_list:
            if i in painters:
                painters[i] +=1
            else:
                painters[i] = 1
                
        return painters
    
if __name__ == "__main__":
    
    N = int(input())
    painting_List = []
    for i in range(N):
        pid = int(input())
        pname = input()
        pprice = int(input())
        ptype = input()
        painting_List.append(Painting(pid,pname,pprice,ptype))
        
    pType = input()
    
    obj1 = ShowRoom(painting_List)
    output1 = obj1.getTotalPaintingPrice(pType)
    output2 = obj1.getMaxPainterName()
    
    
    if output1 == None:
        print("Painting not found")
    else:
        print(output1)
    
    
    names = list(output2.keys())
    paintings = list(output2.values())
    max_paints = max(paintings)
    ind = paintings.index(max_paints)
    print(names[ind])
```
